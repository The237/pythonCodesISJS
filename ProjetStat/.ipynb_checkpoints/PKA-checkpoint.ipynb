## Hypothese de test
les indivudus atteint d'avc ont sysBp significativement different
$H_0$ les taux moyens sont égaux chez les individu ayants eu un AvC et  chez ceux n'ayant pas eu un AVC

# T-Test

```python
from scipy.stats import ttest_ind
```

```python
positive_data.shape
```

```python
negative_data.shape
```

### pour appliquer le test de student il faudrait que le nombre de positif soit egale au nombre de negatif

```python
# pour cela nos allons extraire un échantillons de même taille chez . positifs et chez les 
negative_data.sample(positive_data.shape[0])
```

```python
powo_neg=negative_data.sample(positive_data.shape[0])
```

```python
def t_test(col):
    alpha=0.05
    stat,p=ttest_ind(powo_neg[col].dropna(),positive_data[col].dropna())
    if p<alpha:
        return 'H0 rejetée'
    else:
        return 'H0 pas rejetée'
```

```python
for col in blood:
    print(f'{col:<50}{t_test(col)}')
```
